```python
athletes_df = pd.read_csv("athlete_events.csv")
noc_df = pd.read_csv("noc_regions.csv")

# Merge region (country name) into main dataframe
df = athletes_df.merge(noc_df, how='left', on='NOC')

# Remove entries without a region (country)
df = df.dropna(subset=['region'])

# Step 1: Filter out athletes who didn‚Äôt win a medal
medal_df = df[df['Medal'].notna()]

# Step 2: Count unique athletes sent per Country + Sport
athletes_sent = df.groupby(['region', 'Sport'])['ID'].nunique().reset_index()
athletes_sent.rename(columns={'ID': 'Athletes_Sent'}, inplace=True)

# Step 3: Count medals won per Country + Sport
medals_won = medal_df.groupby(['region', 'Sport'])['Medal'].count().reset_index()
medals_won.rename(columns={'Medal': 'Medals_Won'}, inplace=True)

# Step 4: Merge and calculate medals per athlete
roi_df = pd.merge(athletes_sent, medals_won, on=['region', 'Sport'], how='left')
roi_df['Medals_Won'] = roi_df['Medals_Won'].fillna(0)
roi_df['Medals_per_Athlete'] = roi_df['Medals_Won'] / roi_df['Athletes_Sent']

# Optional: Filter to only show sports with a meaningful number of athletes
roi_df = roi_df[roi_df['Athletes_Sent'] >= 5]

# Step 5: Sort for analysis
roi_df = roi_df.sort_values(by='Medals_per_Athlete', ascending=False)
roi_df
```

##### Question : Which Country has the tallest athletes in Latest Olympics? Display Top 5 (Country Names)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df = pd.read_csv('athlete_events.csv')  
noc = pd.read_csv('noc_regions.csv')

# Merge the country names into the main DataFrame
df = pd.merge(df, noc, how='left', left_on='NOC', right_on='NOC')

# Rename for clarity
df.rename(columns={'region': 'Country'}, inplace=True)
\
df_height = df.dropna(subset=['Height'])

# Filter the latest Olympics year
latest_year = df_height['Year'].max()
latest_olympics = df_height[df_height['Year'] == latest_year]

# Group by Country and compute average height
top5_tallest = (
    latest_olympics.groupby('Country')['Height']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

top5_tallest.columns = ['Country', 'Average_Height']
print(top5_tallest)
```

### This chart reveals which nations consistently produce towering athletes, possibly due to training pipelines, athletic scouting, and sports specialization.
These aren't just tall individuals ‚Äî they're national sports assets.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Define a custom color palette (you can tweak this)
colors = sns.color_palette("viridis", len(top5_tallest))

# Create the plot
plt.figure(figsize=(12, 7))
bars = sns.barplot(
    data=top5_tallest,
    x='Country',
    y='Average_Height',
    palette=colors,
    edgecolor='black'
)

# Add data labels
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(
        f'{height:.1f} cm',
        (bar.get_x() + bar.get_width() / 2, height),
        ha='center', va='bottom',
        fontsize=11, fontweight='bold',
        color='black'
    )

# Titles
plt.title(
    "üèÖ Top 5 Countries with Tallest Athletes (Latest Olympics)",
    fontsize=16, fontweight='bold'
)
plt.suptitle(
    "Average height comparison (in cm) | Which nations stand tallest?",
    fontsize=12, color='gray', y=0.93
)

# Axis labels
plt.xlabel('')
plt.ylabel('Average Height (cm)', fontsize=12)

# Remove top and right borders
sns.despine()

# Tidy layout
plt.tight_layout()
plt.show()
```
